# IPTU Analysis Visualizations

Este notebook demonstra como gerar gráficos a partir dos resultados das análises.


```python
import sys
from pathlib import Path

# Add src to path
sys.path.insert(0, str(Path('..') / 'src'))

from iptu_pipeline.visualizations import IPTUVizualizer
from iptu_pipeline.config import ANALYSIS_OUTPUT_PATH
import matplotlib.pyplot as plt

print("Visualization module loaded successfully!")
```

## 1. Inicializar Visualizador


```python
# Criar visualizador
viz = IPTUVizualizer()

# Verificar se há dados de análise
print(f"Analysis path: {viz.analysis_path}")
print(f"Plots output path: {viz.plots_output_path}")
```

## 2. Gerar Gráficos Individuais


```python
# Volume por ano
fig = viz.plot_volume_by_year(save=False)
if fig:
    plt.show()
```

```python
# Distribuição por tipo de uso
fig = viz.plot_volume_by_type(top_n=10, save=False)
if fig:
    plt.show()
```

```python
# Top bairros
fig = viz.plot_top_neighborhoods(top_n=20, save=False)
if fig:
    plt.show()
```

## 3. Gerar Todos os Gráficos


```python
# Gerar todos os gráficos e salvar
from iptu_pipeline.visualizations import generate_plots_from_analysis_results

plot_files = generate_plots_from_analysis_results()

print(f"\n✓ Gerados {len([f for f in plot_files.keys() if f != 'html_report'])} gráficos")
print(f"\nGráficos salvos em: {plot_files.get('html_report', {}).parent if 'html_report' in plot_files else 'N/A'}")
print(f"\nRelatório HTML: {plot_files.get('html_report', 'N/A')}")
```
