# IPTU Data Pipeline - Execution Notebook

Este notebook demonstra como executar o pipeline completo de processamento de dados IPTU.


```python
import sys
from pathlib import Path

# Add src to path
sys.path.insert(0, str(Path('..') / 'src'))

from iptu_pipeline.orchestration import run_orchestrated_pipeline
from iptu_pipeline.dashboard import IPTUDashboard
from iptu_pipeline.utils.logger import setup_logger

logger = setup_logger('notebook')
```

## 1. Executar Pipeline Completo


```python
# Executar pipeline completo para todos os anos
consolidated_df = run_orchestrated_pipeline(
    years=None,  # None = todos os anos disponíveis
    incremental=False,  # False = processa tudo, True = apenas novos anos
    run_analysis=True  # True = executa análises
)

print(f"Dataset consolidado: {len(consolidated_df):,} linhas, {len(consolidated_df.columns)} colunas")
```

## 2. Gerar Dashboard


```python
# Carregar dados consolidados se necessário
import pandas as pd
from iptu_pipeline.config import CONSOLIDATED_DATA_PATH

if 'consolidated_df' not in locals():
    consolidated_df = pd.read_parquet(CONSOLIDATED_DATA_PATH)

# Criar dashboard
dashboard = IPTUDashboard(df=consolidated_df)

# Gerar dashboard HTML
dashboard_path = dashboard.generate_dashboard_html()
print(f"Dashboard gerado: {dashboard_path}")

# Gerar relatório resumido
report_path = dashboard.generate_summary_report()
print(f"Relatório gerado: {report_path}")
```
